```python
#Answer 2 
import pandas as pd

def clean_data(df):
    """
    Cleans the weather DataFrame by handling missing values appropriately for each column type.
    
    Args:
        df (pd.DataFrame): Raw weather data with potential missing values
        
    Returns:
        pd.DataFrame: Cleaned data with missing values handled
    """

    cleaned_df = df.copy()
    

    numerical_cols = cleaned_df.select_dtypes(include=['int64', 'float64']).columns
    for col in numerical_cols:
        
        cleaned_df[col] = cleaned_df[col].fillna(cleaned_df[col].median())
    
   
    categorical_cols = cleaned_df.select_dtypes(include=['object']).columns
    for col in categorical_cols:
    
        cleaned_df[col] = cleaned_df[col].fillna(cleaned_df[col].mode()[0] if len(cleaned_df[col].mode()) > 0 else 'Unknown')
    
   
    if 'timestamp' in cleaned_df.columns:
        cleaned_df = cleaned_df.sort_values('timestamp')
      
        cleaned_df = cleaned_df.ffill()
    
    return cleaned_df
```

```python
#Anser 3 
def load_data(filepath):
    """Load CSV file into a Pandas DataFrame."""
    try:
        df = pd.read_csv(filepath)
        return df
    except Exception as e:
        print(f"Error loading file: {e}")
        return None

def clean_data(df):
    """Fill missing numeric values with the column mean."""
    return df.fillna(df.mean(numeric_only=True))

def compute_summary(df):
    """Compute and return summary statistics as a dictionary."""
    summary = {
        'mean': df.mean(numeric_only=True).to_dict(),
        'min': df.min(numeric_only=True).to_dict(),
        'max': df.max(numeric_only=True).to_dict()}
    return summary

def save_summary(summary, output_path):
    """Save summary statistics to a text file."""
    try:
        with open(output_path, 'w') as f:
            for stat_name, values in summary.items():
                f.write(f"{stat_name.upper()}:\n")
                for column, value in values.items():
                    f.write(f"  {column}: {value}\n")
                f.write("\n")
        print(f"Summary saved to {output_path}")
    except Exception as e:
        print(f"Error saving summary: {e}")
```
