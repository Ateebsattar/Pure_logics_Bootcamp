```python
# Import the Libaries 
import pandas as pd
import string
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, f1_score
```

```python
# Download NLTK resources outlet 
nltk.download('punkt')
nltk.download('stopwords')
```

```python
# Load the IMDB dataset
df = pd.read_csv("IMDB_dataset.csv")
```

```python
# Preprocessing function
def preprocess(text):
    tokens = word_tokenize(text.lower())  # Lowercase and tokenize
    tokens = [word for word in tokens if word.isalpha()]  # Remove punctuation/numbers
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]  # Remove stopwords
    return ' '.join(tokens)
```

```python
# Apply preprocessing
df['cleaned'] = df['review'].apply(preprocess)
```

```python
# Convert sentiment to binary
df['sentiment'] = df['sentiment'].map({'positive': 1, 'negative': 0})
```

```python
# Split dataset
X_train, X_test, y_train, y_test = train_test_split(df['cleaned'], df['sentiment'], test_size=0.2, random_state=42)
```

```python
# Vectorize text
vectorizer = CountVectorizer()
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)
```

```python
# Train model (you can switch to MultinomialNB() or LinearSVC())
model = LogisticRegression()
model.fit(X_train_vec, y_train)
```

```python
y_pred = model.predict(X_test_vec)
```

```python
# Evaluate
print("Accuracy:", accuracy_score(y_test, y_pred))
print("F1 Score:", f1_score(y_test, y_pred))
```
